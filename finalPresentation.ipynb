### Baseball fields amount


For me, the best neighborhood must have enough amount of baseball fields since I am a big fan of baseball. I really love watching baseball games with friends and families. It is one of the exciting thing for me in my spare time. It is a very personal preference:)))

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("https://data.wprdc.org/dataset/5934b8d1-892f-4fd7-84c1-aa890d3ff529/resource/a61b6617-2f06-4745-8731-165139d41667/download/opendata-pubworks-field-listing-2015.csv")
data.head(10)
```

Above is the dataset **Field Listing 2015**, a list of all baseball fields within the City of Pittsburgh. Although the data is not the newest, it still can reflect something. Then I filter out the unnecessary columns for me and remain the field name and neighborhood.

```python
data_filter = data.filter(regex="Margin")
data.drop(data_filter.columns,axis=1)
data.dropna()
data = data[["Field Name","Neighborhood"]]
data
```

Then group them according to frequency of each neighborhood in descending order.

```python
count = data.groupby(['Neighborhood']).count().sort_values(by=["Field Name"],ascending=False)
print(count)
```

Then plot the bar chart of the above dataframe to make it more intuitive. From the bar chart, we can clearly see the Squirrel Hill South and Brighton Heights have the most number of baseball fields among city of pittsburgh neighborhoods.

```python
count.groupby(['Neighborhood']).median().sort_values("Field Name").plot.bar()
#count.groupby('Neighborhood').agg('Field Name').sum().plot(kind='bar', x='Neighborhood', y='count', rot=90, figsize = (30,20))
```

```python
count.groupby('Neighborhood').agg('Field Name').sum().plot(kind='pie', x='Neighborhood', y='count', rot=90, figsize = (30,12))
```

From the pie chart, we can tell that Squirrel Hill South has 5% of the baseball fields out of all the baseball fields in city of pittsburgh

### The number of playround equipments in a neighborhood

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Another important thing for a good neighborhood is to have many playround equipments. Children and elderly people can have fun in playround and this helps families to know other neighbors. I am using the dataset of Playground Equipment from WPRDC, a listing of all playground equipment maintained by the city, including a description, the manufacturer, included safety surface and whether or not it is ADA accessible. The first step is to remove all columnes that are not useful for me.

```python
data = pd.read_csv("https://data.wprdc.org/datastore/dump/e39ef76e-0a11-47c8-a86f-a37f55db7a2b")
data.head(10)
```

Below is the way to filter out all columns except name of playground and its neighborhood location.

```python
data_filter = data.filter(regex="Margin")
data.drop(data_filter.columns,axis=1)
data.dropna()
data = data[["name","neighborhood"]]
data
```

Then I need to get the occurences of each neighbood displayed in this list. And I found that Squirrel Hill South has the most number of playgrounds.

```python
count = data.groupby(['neighborhood']).count().sort_values(by=["name"],ascending=False)
print(count)
```

Then I make a bar chart in ascending order to make the data more intuitive to analyze. 

```python
count.groupby(['neighborhood']).median().sort_values("name").plot.bar()
plt.rcParams['figure.figsize'] = [50, 50]
#count.groupby('INCIDENTNEIGHBORHOOD').agg('PK').sum().plot(kind='bar', x='Neighborhood', y='count', rot=90, figsize = (30,20))
```

The I narrow down the bar chart to aound top 20 neighborhoods in the left of previous bar chart

```python
count.groupby(['neighborhood']).median()[40:].sort_values("name").plot.bar()
```

### Public art counts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

I am using the dataset **City of Pittsburgh Public Art** which shows all public art projects in the City of Pittsburgh, including type of art, artist's name, and location.

```python
data = pd.read_csv("https://data.wprdc.org/datastore/dump/00d74e83-8a23-486e-841b-286e1332a151")
data.head(10)
```

1. remove unuseful columns and just remain the title and neighborhood

```python
data_filter = data.filter(regex="Margin")
data.drop(data_filter.columns,axis=1)
data.dropna()
data = data[["title","art_type","neighborhood"]]
data
```

```python
#data['neighborhood'].value_counts()[:10].index.tolist()
```

Below is to get the occurences of each neighborhood in a descending order and we can see that Squirrel Hill South is the top2 neighborhood of have the most number of art places.

```python
count = data.groupby(['neighborhood']).count().sort_values(by=["title"],ascending=False)
print(count)
```

Then I plot the bar chart in ascending order

```python
#count.groupby('neighborhood').agg('title').sum().plot(kind='pie', x='Neighborhood', y='count', rot=90, figsize = (30,20))
count.groupby(['neighborhood']).median().sort_values("title").plot.bar()
```

Then I narrow down the bar chart to see statistcs about several top neighborhoods

```python
count.groupby(['neighborhood']).median()[35:].sort_values("title").plot.bar()
```

```python
The bar plot shows us in a more intuitive way of the top neighborhoods for having the number of art places.
```

```python
plot = count.plot.pie(subplots=True, figsize=(35,40))
```

From the pie, we can roughly know the number of art places in a given community as a percentage of the tota, Like the **squirell hill south** is roughly 10%.
